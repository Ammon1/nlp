```python
def split_into_sentences(text):
    sentences = text.split("\n")
    sentences = [s.strip() for s in sentences]
    return sentences

def download_article(link):
    article = Article(link)
    article.download()
    try:
        article.parse()
    except:
        return 'nic'
    return article.text
```

```python
def cleaning(doc):
    txt = [token.lemma_ for token in doc if not token.is_stop]
    if len(txt) > 2:
        return ' '.join(txt)
```
